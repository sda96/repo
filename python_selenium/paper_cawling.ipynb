```python
from tqdm import tqdm
from selenium.common.exceptions import UnexpectedAlertPresentException
from selenium.common.exceptions import NoSuchElementException
```

```python
d = pd.DataFrame()
#131 , 1956
```

```python
for k in tqdm(range(1957,len(part_url))):
    try:
        driver.get(part_url.urls[k])
        body = driver.find_element_by_css_selector("#rndViewForm > div > div > div.notice_view")
        title = body.find_element_by_css_selector("div.ditail_tit").text
        depart = body.find_element_by_css_selector("li:nth-child(3)").text
        contents = body.find_element_by_css_selector("div.notice_cont").text
        files = driver.find_element_by_css_selector("#rndViewForm > div > div > div.notice_view > div.summary_file > ul")
        download = files.find_elements_by_css_selector("li > a")
        text = []
        for i in range(0,len(download)):
            do_te = download[i].text
            text.append(do_te)
        matching = [[idx,s] for idx,s in enumerate(text) if "공고" in s] 
        try:
            for i in range(len(matching[0])+1):
                download[matching[i][0]].click()
                time.sleep(3)
                try:
                    filename = matching[i][1]
                    contents2 = subprocess.check_output("hwp5txt"+" "+'\"'+"%s"%filename+'\"',shell=True,encoding="utf-8")
                except:
                    contents2 = "-"
        except:
            pass
        time.sleep(2)
        data = pd.DataFrame({"제목":title,"부처":depart,"내용":contents,"다운로드":matching,"파일내용":contents2})
        d = d.append(data)
    except UnexpectedAlertPresentException:
        driver.refresh()
        time.sleep(10)
    except NoSuchElementException:
        pass
d.to_excel("./자료.xlsx",sheet_name = "Sheet1")
```

```python
def clean_text(text):
    import re
    text = re.sub(r'[\n\t\r]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    text = re.sub('<.+?>', ' ', text)
    cleaned = text.strip()
    return(cleaned)
```
